```python
# Factorial of a number using loop

n = int(input("Enter a number :"))
if n<0:
  print("Factorial of a negative number doesn't exist !!!")
else:
  factorial=1
  for i in range(1, n + 1):
    factorial *= i
  print(factorial)
```

```python
def calculator(numbers):
    print("\n--- Menu ---")
    print("1. Addition\n2. Subtraction\n3. Multiplication\n4. Division\n5. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        result = sum(numbers)
        print(f"Total Sum: {result}")

    elif choice == 2:
        # Subtracts all subsequent numbers from the first one
        result = numbers[0]
        for num in numbers[1:]:
            result -= num
        print(f"Total Subtraction: {result}")

    elif choice == 3:
        result = 1
        for num in numbers:
            result *= num
        print(f"Total Multiplication: {result}")

    elif choice == 4:
        result = numbers[0]
        try:
            for num in numbers[1:]:
                result /= num
            print(f"Total Division: {result}")
        except ZeroDivisionError:
            print("Error: Cannot divide by zero.")

    elif choice == 5:
        print("Exiting...")
        return

    else:
        print("Invalid choice!")

# --- Main Execution ---
nums = []
n = int(input("How many numbers do you want to use? "))

for i in range(n):
    val = int(input(f"Enter number {i + 1}: "))
    nums.append(val)

calculator(nums)
```

```python
def check_element(any_tuple, element):
  if element in any_tuple:
    print(element, "is present in the tuple")
  else:
    print(element, "is not present in the tuple")


any_tuple = ()
n = int(input("Enter the number of elements in the tuple :"))
for i in range(n):
  element = int(input("Enter the element :"))
  any_tuple += (element,)
element = int(input("Enter the number to be searched :"))
check_element(any_tuple, element)
```
